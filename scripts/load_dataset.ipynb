```python
from datasets import load_dataset

ds = load_dataset("pminervini/HaluEval", "qa_samples")

# save the dataset to a json file
ds.to_json("data/halueval/qa_samples.json")

# load the dataset from the json file
ds = load_dataset("json", data_files="qa_samples.json")

```

```python
# split to 80% and 20%
split_ds = ds['data'].train_test_split(test_size=0.2)
train_ds = split_ds['train']
test_ds = split_ds['test']

# save the train and test datasets to a json file
train_ds.to_json("data/halueval/train_qa_samples.json")
test_ds.to_json("data/halueval/test_qa_samples.json")
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-7B-Instruct")
model = AutoModelForCausalLM.from_pretrained(
    "Qwen/Qwen2.5-7B-Instruct",
    device_map="auto",  # Automatically distribute model across available GPUs
    # torch_dtype=torch.float16,  # Use float16 to save memory
)

messages = [
    {"role": "user", "content": "Who are you?"},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to("cuda")  # Move inputs to GPU

outputs = model.generate(**inputs, max_new_tokens=40)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```
